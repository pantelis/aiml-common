```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate bivariate Gaussian data
mean = [0, 0]
cov = [[1, 0], [0, 1]]  # Diagonal covariance, independent distributions
z  = np.random.multivariate_normal(mean, cov, 1000)

# Step 2: Apply the transformation g(x)
# def transform(z):
#     x = np.zeros(z.shape)
#     x[0,] = (z[0,] / np.linalg.norm(z))
#     x[1,] = (z[1,] / np.linalg.norm(z))
#     return x

def transform(z):
    x = np.zeros(z.shape)
    z_l1_norm =  np.linalg.norm(z, ord=1, axis=1, keepdims=True)
    x = z/10 + z / z_l1_norm
    return x


x = transform(z)

# Step 3: Plot the original and transformed data
plt.figure(figsize=(12, 6))

# Original data plot
plt.subplot(1, 2, 1)
plt.plot(z[:,0], z[:,1], 'o')
plt.title('Original Bivariate Gaussian')
plt.xlabel('z_1')
plt.ylabel('z_2')
plt.axis('equal')

# Transformed data plot
plt.subplot(1, 2, 2)
plt.plot(x[0,], x[1,], 'o')
plt.title('Transformed Distribution')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.axis('equal')

plt.show()
```
